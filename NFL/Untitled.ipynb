```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
def scrape_passing():

    BASE_URL = 'http://www.espn.com/nfl/weekly/leaders/_/week/{}/seasontype/2/type/passing'
    
    Passing_df = pd.DataFrame()
    Weekly_df = pd.DataFrame()

    for i in range(1,19):
        print(i)
        URL = BASE_URL.format(i)
        page = requests.get(URL)
        if page.ok:
            soup = BeautifulSoup(page.content, "html.parser")
            table = soup.find_all('table')
            df = pd.read_html(str(table))[11]
            time.sleep(1)
        col_names = list(df.iloc[1, :])
        df = df.iloc[2: , :]
        df.columns = col_names
        df = df.reset_index(drop = True)
        
        # Add Week Number
        df["Week"] = i
        
        # Calculate Fantasy Points
        df['PASSING_FPTS'] = (df['TD'].astype(int) * 4) + (df['YDS'].astype(int) / 25) + (df['INT'].astype(int) * -2)
        
        # Add Result
        df[['GAME_RESULT', 'FINAL_SCORE', 'vs', 'OPP_TEAM']] = df['RESULT'].str.split(' ', 3, expand=True)
        df = df.drop(['RK', 'RESULT', 'vs'], axis = 1)
        
        # Add to total df
        Passing_df = pd.concat([Passing_df, df], ignore_index=True)
        
        if i == 1:
            columns = ['PLAYER', 'WEEK 1']
            Weekly_df = df[['PLAYER', 'PASSING_FPTS']]
            Weekly_df.columns = columns
        else:
            temp = df[['PLAYER', 'PASSING_FPTS']]
            temp.columns = ['PLAYER', 'WEEK {}'.format(i)]
            Weekly_df = Weekly_df.merge(temp, on='PLAYER', how='outer')
    return df, Passing_df, Weekly_df
    
df1, Passing_df, Weekly_df_pass = scrape_passing()
```
